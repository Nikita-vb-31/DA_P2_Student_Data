```python
#Building ML Model
from sklearn.model_selection import train_test_split
```

```python
#Identifying X(independent) and Y(dependent) variables
Y=data['TotalScore']
X=data.drop(['TotalScore','MathScore','ReadingScore','WritingScore'],axis=1)
```

```python
data.head()
```

```python
X.head()
```

```python
Y.head()
```

```python
#Observe that there are many categorical variables in the X(independent) set of data.
#To build the model we have to encode all those variables.
from sklearn.preprocessing import LabelEncoder
encode=LabelEncoder()
```

```python
#encoding gender 
X['Gender_encoded']=encode.fit_transform(X['Gender'])
```

```python
X.head()
```

```python
#encoding other variables like EthnicGroup ParentEduc LunchType TestPrep
X['EthnicGroup_encoded']=encode.fit_transform(X['EthnicGroup'])
X['ParentEduc_encoded']=encode.fit_transform(X['ParentEduc'])
X['LunchType_encoded']=encode.fit_transform(X['LunchType'])
X['TestPrep_encoded']=encode.fit_transform(X['TestPrep'])
```

```python
X.head()
```

```python
#droping all the categorical variables and also sl.no. as it would not contribute much for our model 
X=X.drop(['SL.No.','Gender','EthnicGroup', 'ParentEduc', 'LunchType', 'TestPrep'],axis=1)
```

```python
X.head()
```

```python
#splitting the dataset 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3)
```

```python
#Standardizing train and test splits
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_train
```

```python
x_test=scaler.transform(x_test)
```

```python
x_test
```

```python
#Model Training 
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
#prediction with test data
regression_pred=regression.predict(x_test)
```

```python
print(regression_pred)
```

```python
#scatter plot for prediction
plt.scatter(y_test,regression_pred,color='orange')
plt.xlabel('Actual TotalScore') 
plt.ylabel('Predicted TotalScore') 
plt.title("Actual vs Predicted TotalScoreScores")
```

```python
residuals=y_test-regression_pred
```

```python
print(residuals)
```

```python
##performance matrix

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
print(mean_squared_error(y_test,regression_pred))
print(mean_absolute_error(y_test,regression_pred))
print(np.sqrt(mean_squared_error(y_test,regression_pred)))
```

```python
df_results = pd.DataFrame({'Actual': y_test, 'Predicted': regression_pred})
```
